#### Regression Problem on the same dataset


```python
import pandas as pd
import tensorflow as tf
import numpy as np
import tensorboard as tb
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,LabelEncoder, OneHotEncoder
import pickle
```

```python
data  = pd.read_csv('Churn_Modelling.csv')
data.head()
```

```python
data = data.drop(['RowNumber','CustomerId','Surname'],axis=1)
data.head()
```

```python
label_encoder_gender = LabelEncoder()
data['Gender'] = label_encoder_gender.fit_transform(data['Gender'])
data.head(10)
```

```python
# One hot encodind for geography 
one_hot_encoder = OneHotEncoder()
geo_encoder = one_hot_encoder.fit_transform(data[['Geography']])
geo_encoder
```

```python
encoded_df = pd.DataFrame(geo_encoder.toarray(),columns=one_hot_encoder.get_feature_names_out(['Geography']))
encoded_df
```

```python
data = pd.concat([data.drop('Geography',axis=1),encoded_df],axis=1)
data.head(30)
```

```python
y = data['EstimatedSalary']
x = data.drop('EstimatedSalary',axis=1)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=45,test_size=0.2)
```

```python
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

model = Sequential([
    Dense(64,activation = 'relu',input_shape = (x_train.shape[1],)),
    Dense(32,activation = 'relu'),
    Dense(8,activation='relu'),
    Dense(1)
])

model.compile(optimizer='adam',loss='mean_squared_error',metrics=['mse'])
model.summary()
```

```python
from tensorflow.keras.callbacks import EarlyStopping
early_stopping = EarlyStopping(patience = 10 , monitor='val_loss', restore_best_weights = True)
```

```python
history = model.fit(
    x_train,y_train,
    validation_data = (x_test,y_test),
    epochs = 100,
    callbacks = [early_stopping]
)
```

```python
test_loss,test_mse = model.evaluate(x_test,y_test)
print(f'TEST MSE : {test_mse}, TEST LOSS : {test_loss}')
```
